7. Write a simpler bisection-utilizing algorithm that computes the fourth root of a given
number. Indent your code to explain what each step is doing. Multiple ways of coding
this.

````
#y**4 = x <-> y = 4th root of x e.g. y = 3 and x= 81

number = 81   # fourth root = 3
low = 0          # low = 0 since lowest square root value is greater than zero. 
high = number    # Highest value of square root cant be more than the number itself. 
guess = number/2 # Take an initial guess in the midpoint of high and low
epsilon = 0.01   # The inaccuracy that we tolerate. Higher the epsilon means lower accuracy result/guess.
num_guesses = 0  # num_guesses keeps track of the number of guesses taken to reach result. 

while ( abs(guess**4 - number) >= epsilon):  # The loop runs as long as the guess is not within epsilon (0.01) distance of number.

    if guess**4 > number:                    # If our guess for the square root is too high. Example: we guess 7 is sqrt(25). 
        high = guess                         # Then the higher bound of our next guess is our previous guess. Example: high = 7 (down from high = 25)
    else:                                    # If our guess is too low. Example we guess 3 for sqrt(25).
        low = guess                          # Then the lower bound of our next guess is our previous guess. Example: low = 3 (up from low = 0)

    guess = (high + low)/2                   # Since now we reduced our guess space, we make our next guess by taking the midpoint of new low or high ...
                                             # ... Example: previous geuss was 7 for sqrt(25). hence new guess is:( for low is 0 and high is 7) 3.5. 

    num_guesses = num_guesses +  1                         # Record that a new guess has been made.
print(guess, " ", "iterations for convergence: ", num_guesses) # Print that information to user.

                        # return guess. 
````

8. Convert your code for question 7 into a function. Check your code by computing square
roots for: 5, 125, 40.

````
def  fourth_root (n):
  epsilon = 0.01
  num_guesses = 0
  low = 0
  high = n
  guess = (high + low)/2.0

  while ( abs(guess**4 - n) >= epsilon):  # The loop runs as long as the guess is not within epsilon (0.01) distance of number.

    if guess**4 > n:                    # If our guess for the square root is too high. Example: we guess 7 is sqrt(25). 
        high = guess                         # Then the higher bound of our next guess is our previous guess. Example: high = 7 (down from high = 25)
    else:                                    # If our guess is too low. Example we guess 3 for sqrt(25).
        low = guess                          # Then the lower bound of our next guess is our previous guess. Example: low = 3 (up from low = 0)

    guess = (high + low)/2                   # Since now we reduced our guess space, we make our next guess by taking the midpoint of new low or high ...
                                             # ... Example: previous geuss was 7 for sqrt(25). hence new guess is:( for low is 0 and high is 7) 3.5. 

    num_guesses = num_guesses +  1                         # Record that a new guess has been made.
  print(guess, " ", "iterations for convergence: ", num_guesses)

  return guess
````

````
print(fourth_root(5))
print(fourth_root(125))
print(fourth_root(40))
````

Machine learning models require all input and output variables to be numeric.

This means that if your data contains categorical data, you must encode it to numbers before you can fit and evaluate a model.

The two most popular techniques are an **Ordinal Encoding** and a **One-Hot Encoding**.

Categorical data are variables that contain label values rather than numeric values.

The number of possible values is often limited to a fixed set.

Categorical variables are often called nominal. <br>


In *ordinal encoding*, each unique category value is assigned an integer value.

For example, “red” is 1, “green” is 2, and “blue” is 3.

This is called an ordinal encoding or an integer encoding and is easily reversible. Often, integer values starting at zero are used.
